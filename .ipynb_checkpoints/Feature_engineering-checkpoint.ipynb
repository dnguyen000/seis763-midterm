```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ta # for technical analysis like RSI,MACD
import warnings
```

```python
df_spy  = pd.read_csv("data/SPY_raw.csv",  index_col="date", parse_dates=True)
df_tsla = pd.read_csv("data/TSLA_raw.csv", index_col="date", parse_dates=True)
df_vix  = pd.read_csv("data/VIX_raw.csv",  index_col="date", parse_dates=True)

print("SPY: ",  df_spy.shape)
print("TSLA:", df_tsla.shape)
print("VIX: ",  df_vix.shape)
```

```python
def build_features(df, df_vix):
    
    close = df["adj_close"]
    # price features
    df["daily_return"]   = close.pct_change()                      # % change from yesterday to today
    df["weekly_return"]  = close.pct_change(5)                     # % change from 5 trading days ago (1 business week)
    df["ma_7"]           = close.rolling(7).mean()                 # average closing price over last 7 trading days
    df["ma_21"]          = close.rolling(21).mean()                # average closing price over last 21 trading days (1 month)
    df["ma_cross"]       = df["ma_7"] - df["ma_21"]                # positive = short term above long term (bullish), negative = bearish
    df["dist_from_ma21"] = (close - df["ma_21"]) / df["ma_21"]    # how far price is from monthly average, as a %
    df["daily_range"]    = (df["high"] - df["low"]) / close        # today's price swing top to bottom relative to close
    
    # momentum features
    df["rsi_14"]      = ta.momentum.RSIIndicator(close=close, window=14).rsi()  # momentum 0-100, above 70 = overbought, below 30 = oversold for 14 trading days

    macd              = ta.trend.MACD(close=close)
    df["macd"]        = macd.macd()                                              # difference between 12 and 26 day moving average
    df["macd_signal"] = macd.macd_signal()                                       # smoothed 9 day average of macd line
    df["macd_hist"]   = macd.macd_diff()                                         # macd minus signal, positive = bullish momentum, negative = bearish

    bb                = ta.volatility.BollingerBands(close=close, window=20, window_dev=2)
    df["bb_position"] = (close - bb.bollinger_lband()) / (bb.bollinger_hband() - bb.bollinger_lband())  # 0 = at lower band, 1 = at upper band, 0.5 = middle

    df["volatility_7"]  = df["daily_return"].rolling(7).std()                   # std of daily returns over 7 days, high = volatile period
    df["volatility_20"] = df["daily_return"].rolling(20).std()                  # std of daily returns over 20 days, smoother longer term volatility
    
    # volume features
    df["volume_change"] = df["volume"].pct_change()                              # % change in volume from yesterday to today
    df["volume_ma20"]   = df["volume"].rolling(20).mean()                        # average volume over last 20 trading days
    df["volume_ratio"]  = df["volume"] / df["volume_ma20"]                       # today's volume vs 20 day average, above 1 = unusually high volume

    
    # lagged returns
    df["lag_return_1"] = df["daily_return"].shift(1)                             # what was the return 1 day ago
    df["lag_return_3"] = df["daily_return"].shift(3)                             # what was the return 3 days ago
    df["lag_return_5"] = df["daily_return"].shift(5)                             # what was the return 5 days ago

    
    # seasonal features
    df["month"]   = df.index.month                                               # month number 1-12
    df["quarter"] = df.index.quarter                                             # quarter number 1-4

    df["season_num"] = df["month"].map({
        12: 0, 1: 0, 2: 0,                                                      # winter = 0
        3: 1,  4: 1, 5: 1,                                                      # spring = 1
        6: 2,  7: 2, 8: 2,                                                      # summer = 2
        9: 3, 10: 3, 11: 3                                                      # fall = 3
    })
    df["season"] = df["season_num"].map({0: "Winter", 1: "Spring", 2: "Summer", 3: "Fall"})  # season name

    df["is_earnings_week"] = df.index.map(
        lambda dt: 1 if dt.month in [1, 4, 7, 10] and dt.day >= (__import__('calendar').monthrange(dt.year, dt.month)[1] - 14) else 0
    )                                                                            # 1 if last 2 weeks of Jan, Apr, Jul, Oct (earnings season)

    
    # vix features
    df["vix"]            = df_vix["close"].reindex(df.index, method="ffill")    # raw vix value for that day
    df["is_major_event"] = (df["vix"] > 30).astype(int)                         # 1 if vix above 30, market stress event

    # targets
    df["target_direction"] = (close.shift(-1) > close).astype(int)              # 1 if tomorrow is higher than today (UP), 0 if DOWN
    df["target_return"]    = (close.shift(-5) - close) / close                  # exact % price change over next 5 trading days
    
    return df
```

```python
df_spy  = build_features(df_spy,  df_vix)
df_tsla = build_features(df_tsla, df_vix)

print("SPY: ",  df_spy.shape)
print("TSLA:", df_tsla.shape)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
fig.suptitle("Target Balance — UP vs DOWN Days", fontsize=14, y=1.02)

for ax, df, ticker in zip(axes, [df_spy, df_tsla], ["SPY", "TSLA"]):
    counts = df["target_direction"].value_counts().sort_index()
    bars = ax.bar(["DOWN (0)", "UP (1)"], counts.values, color=["red", "green"], alpha=0.8)
    for bar, val in zip(bars, counts.values):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5,
                f"{val}\n({val/len(df)*100:.1f}%)", ha="center", fontsize=11)
    ax.set_title(f"{ticker}\nUP={counts.get(1,0)/len(df)*100:.1f}%  DOWN={counts.get(0,0)/len(df)*100:.1f}%")
    ax.set_ylabel("Count")
    ax.set_ylim(0, counts.max() * 1.2)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle("RSI-14 Distribution", fontsize=14)

for ax, df, ticker, color in zip(axes, [df_spy, df_tsla], ["SPY", "TSLA"], ["blue", "red"]):
    ax.hist(df["rsi_14"], bins=50, color=color, alpha=0.7)
    ax.axvline(x=70, color="red",   linestyle="--", linewidth=1.5, label="Overbought (70)")
    ax.axvline(x=30, color="green", linestyle="--", linewidth=1.5, label="Oversold (30)")
    ax.axvline(x=50, color="gray",  linestyle=":",  linewidth=1,   label="Neutral (50)")
    ax.set_title(f"{ticker}")
    ax.set_xlabel("RSI Value")
    ax.set_ylabel("Frequency")
    ax.legend()

plt.tight_layout()
plt.show()
```

## RSI-14 Distribution

RSI (Relative Strength Index) measures momentum on a 0-100 scale based on the last 14 trading days. It shows whether a stock is overbought or oversold relative to its recent history.

| Zone | Range | Meaning |
|---|---|---|
| Oversold | Below 30 | Stock has been beaten down, may bounce |
| Neutral | 30 to 70 | Normal trading range |
| Overbought | Above 70 | Stock has run up too far, may pull back |

### SPY
- Distribution is skewed right — most days cluster between 50-70
- Rarely goes below 30 — SPY almost never gets that beaten down
- Frequently pushes above 70 — makes sense since SPY trends upward and spends a lot of time in overbought territory
- RSI below 30 for SPY is a rare and meaningful signal when it does occur

### TSLA
- Much more spread out, closer to a normal bell curve
- Touches both extremes regularly — goes below 30 and above 70 far more often than SPY
- Centered around 50 — TSLA swings both directions equally with no strong directional bias
- RSI extremes are more common for TSLA so they carry less weight as a signal on their own

### What This Means For Modeling
- RSI behaves differently for each ticker — another reason to train SPY and TSLA separately
- For SPY, an RSI below 30 is a rare event worth paying attention to
- For TSLA, RSI extremes are routine and the model will need other features to confirm the signal

```python
import seaborn as sns

model_features = [
    "daily_return", "weekly_return", "ma_cross", "dist_from_ma21", "daily_range",
    "rsi_14", "macd_hist", "bb_position", "volatility_7", "volatility_20",
    "volume_change", "volume_ratio",
    "lag_return_1", "lag_return_3", "lag_return_5",
    "season_num", "month", "quarter", "is_earnings_week",
    "vix", "is_major_event",
    "target_direction"
]

fig, axes = plt.subplots(1, 2, figsize=(20, 16))
fig.suptitle("Feature Correlation Heatmap", fontsize=14)

for ax, df, ticker in zip(axes, [df_spy, df_tsla], ["SPY", "TSLA"]):
    corr = df[model_features].corr()
    sns.heatmap(corr, ax=ax, cmap="coolwarm", center=0, vmin=-1, vmax=1,
                annot=True, fmt=".2f", annot_kws={"size": 7},
                linewidths=0.3)
    ax.set_title(ticker, fontsize=12)
    ax.tick_params(axis="x", rotation=45, labelsize=8)
    ax.tick_params(axis="y", rotation=0,  labelsize=8)

plt.tight_layout()
plt.show()
```

## Feature Correlation Heatmap

The heatmap shows how correlated every feature is with every other feature.

| Color | Meaning |
|---|---|
| Dark red | Strong positive correlation (+1) |
| Dark blue | Strong negative correlation (-1) |
| White / light | No correlation (0) |

The diagonal is always 1.0 — every feature is perfectly correlated with itself.

### Key Findings

**Redundant features (highly correlated with each other):**
- `ma_cross` and `dist_from_ma21` are almost measuring the same thing
- `month`, `quarter`, `season_num` are all just different ways of expressing calendar position
- `volatility_7` and `volatility_20` move together since short and long volatility are related
- `vix` and `is_major_event` are obviously correlated since one is derived from the other

**Correlation with target_direction (bottom row):**
- Almost all features show light colors against `target_direction`
- No single feature strongly predicts direction on its own
- This is expected — if one feature had a 0.8 correlation with the target you would not need ML at all

**SPY vs TSLA differences:**
- TSLA shows stronger correlations between price features — its trends are more dramatic so momentum indicators move more together
- SPY correlations are more diffuse and spread out

### What This Means For Modeling
- Consider dropping one of `month`, `quarter`, or `season_num` since they carry the same information
- Keep both `vix` and `is_major_event` despite their correlation — they capture different signals
- No single feature dominates — the model needs all of them working together

```python
fig, axes = plt.subplots(3, 2, figsize=(16, 12))
fig.suptitle("Key Features Over Time", fontsize=14)

features = ["ma_cross", "bb_position", "rsi_14"]
labels   = ["MA Cross (MA7 - MA21)", "Bollinger Band Position", "RSI-14"]

for row, (feat, label) in enumerate(zip(features, labels)):
    for col, (df, ticker, color) in enumerate(zip([df_spy, df_tsla], ["SPY", "TSLA"], ["blue", "red"])):
        ax = axes[row][col]
        ax.plot(df.index, df[feat], color=color, linewidth=0.8, alpha=0.8)
        ax.axhline(y=0,  color="black", linestyle="--", linewidth=0.8)
        ax.set_title(f"{ticker} — {label}")
        ax.set_ylabel(label)

plt.tight_layout()
plt.show()
```

## Key Features Over Time

### MA Cross (MA7 - MA21)
- SPY stays close to zero with small consistent swings throughout the 10 years
- The big dip in 2020 is the COVID crash where the short term average fell hard below the long term average
- TSLA is flat near zero from 2015-2019 when the price was low, then explodes in 2020-2021 with massive swings of +/-40 points as the stock went parabolic
- Above zero = short term trend is bullish. Below zero = bearish.

### Bollinger Band Position
- Both tickers oscillate rapidly between 0 and 1 — price constantly bounces inside the bands
- Values above 1 or below 0 mean price broke outside the bands, an extreme move
- SPY breaks outside the bands less often than TSLA, confirming SPY is more contained and predictable

### RSI-14
- SPY RSI rarely drops below 30 — only visibly during COVID 2020 and a couple other brief moments
- TSLA RSI touches both extremes much more regularly, going below 30 and above 70 frequently
- Both spend most of their time between 40-70 which matches the RSI distribution histogram

### Overall Takeaway
All three features clearly capture the COVID crash in 2020 — MA cross dips hard, Bollinger Band position spikes, and RSI crashes at the same time. This confirms these features carry real information about market stress periods and will give the model meaningful signals to learn from.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle("Target Return Distribution — 5 Day Forward Return", fontsize=14)

for ax, df, ticker, color in zip(axes, [df_spy, df_tsla], ["SPY", "TSLA"], ["blue", "red"]):
    ax.hist(df["target_return"] * 100, bins=80, color=color, alpha=0.7)
    ax.axvline(x=0, color="black", linestyle="--", linewidth=1, label="Zero")
    ax.axvline(x=df["target_return"].mean() * 100, color="green", linestyle="--", linewidth=1, label=f"Mean={df['target_return'].mean()*100:.2f}%")
    ax.set_title(f"{ticker}")
    ax.set_xlabel("5 Day Forward Return (%)")
    ax.set_ylabel("Frequency")
    ax.legend()

plt.tight_layout()
plt.show()
```

## Target Return Distribution — 5 Day Forward Return

This is the regression target. It shows the distribution of actual % price change over the next 5 trading days across all 2,515 rows.

### SPY
- Tight distribution, most 5 day returns fall between -5% and +5%
- Mean of +0.28% — slightly positive, markets drift upward over time
- Almost perfectly normal bell curve — predictable and well behaved
- Extreme moves beyond -10% or +10% are very rare

### TSLA
- Much wider and flatter distribution, regularly swings -20% to +20% over 5 days
- Mean of +1.03% — higher average return than SPY but with much higher risk
- Fatter tails on both sides — extreme moves happen regularly
- X-axis extends to +60% — TSLA had 5 day periods where it gained 60% during the 2020 boom

### What This Means For Modeling
- SPY regression will be easier — returns are tight and centered, less noise for the model
- TSLA regression will be much harder — the wide spread means even a good model will have large prediction errors
- Expect RMSE to be significantly higher for TSLA than SPY — this is not a failure, it is the nature of the stock
- A model that predicts TSLA direction correctly 55% of the time is actually a strong result given this distribution
